```python
words = open('names.txt', 'r').read().splitlines()
```

```python
words[:10]
```

```python
b = {}
for w in words:
    chs = ['<S>'] + list(w) + ['<E>']
    
    for ch1, ch2 in zip(chs, chs[1:]):
        # count (ch1, ch2) in b
        if (ch1, ch2) in b:
            b[(ch1, ch2)] += 1
        else:
            b[(ch1, ch2)] = 1
            
sorted(b.items(), key=lambda x: x[1])
            
b
```

```python
chars = sorted(list(set(''.join(words))))
stoi = {s: i for i, s in enumerate(chars)}
stoi['<S>'] = len(stoi)
stoi['<E>'] = len(stoi)
```
